```julia
function Order_Acceptance(Seq::Vector{Int}, r::Vector{Int64}, p::Vector{Int64}, d::Vector{Int64}, d_bar::Vector{Int64}, 
        e::Vector{Int64}, w::Vector{Float64}, S::Matrix{Int64})
    n = length(r)
    Tmax = maximum(d_bar)
    F = zeros(n, Tmax)  #Rewards
    P = zeros(Int, n, Tmax)  #last scheduled job (for sequence dependent setup time)
    T = zeros(Int, n, Tmax)  #previous time (in order to extract the final sequence at the end)
    for k2 = 1:n
        for t = r[Seq[k2]]:Tmax - p[Seq[k2]] 
            t_k = t + S[1, Seq[k2]+1] + p[Seq[k2]] 
            Tardiness = max(0, t_k-d[Seq[k2]])
            Reward = max(0.0, e[Seq[k2]] - Tardiness*w[Seq[k2]])
            if Reward > 0
                F[k2,t_k] = Reward
                P[k2,t_k] = Seq[k2]
            end
        end
    end

    for k1 = 1:n-1
        for k2 = k1+1:n
            for t = r[Seq[k2]]:Tmax-p[Seq[k2]] 
    #             updated = false
                t_k = t + S[P[k1,t]+1, Seq[k2]+1] + p[Seq[k2]] 
                if t_k < Tmax
                    Tardiness = max(0, t_k-d[Seq[k2]])
                    Reward = max(0.0, e[Seq[k2]] - Tardiness*w[Seq[k2]])
                    if F[k1,t] + Reward > F[k2,t_k]
                        F[k2,t_k] = F[k1,t] + Reward
                        P[k2,t_k] = Seq[k2]
                        T[k2,t_k] = t
                        updated = true
                    end
                end
    #             if !updated            
            end
        end
    end
    return maximum(F)
end
```

```julia
function Order_Acceptance(Seq::Vector{Int}, r::Vector{Int64}, p::Vector{Int64}, d::Vector{Int64}, d_bar::Vector{Int64}, 
        e::Vector{Int64}, w::Vector{Float64}, S::Matrix{Int64})
    n = length(r)
    Tmax = maximum(d_bar)
    F = zeros(n, n)  #Rewards
    M = zeros(Int, n, n)  #previous scheduled job (for sequence dependent setup time)
    T = zeros(Int, n, n)  #finishing time 
    for i = 1:n
        t_k = r[Seq[i]] + S[1, Seq[i]+1] + p[Seq[i]] 
        if t_k < d_bar[i]
            Tardiness = max(0, t_k-d[Seq[i]])
            Reward = max(0.0, e[Seq[i]] - Tardiness*w[Seq[i]])
            if Reward > 0
                F[1,i] = Reward
                T[1,i] = t_k
            end
        end
    end

    for k = 2:n-1
        for i = k:n
            t_k = r[Seq[i]] + S[1, Seq[i]+1] + p[Seq[i]] 
            if t_k < d_bar[i]
                Tardiness = max(0, t_k-d[Seq[i]])
                Reward = max(0.0, e[Seq[i]] - Tardiness*w[Seq[i]])
                if Reward > 0
                    F[1,i] = Reward
                    T[1,i] = t_k
                end
            end
        end
    end
    return maximum(F)
end
```

```julia
function Order_Acceptance(Seq::Vector{Int}, r::Vector{Int64}, p::Vector{Int64}, d::Vector{Int64}, d_bar::Vector{Int64}, e::Vector{Int64}, w::Vector{Float64}, S::Matrix{Int64})
    n = length(r)
    T = maximum(d_bar)
    R = zeros(n, T)  #Rewards
    M = zeros(Int, n, T)  #last scheduled job (for sequence dependent setup time)
    P = zeros(Int, n, T)  #previous time (in order to extract the final sequence at the end)

    for t=1:T
        t_k = r[Seq[1]] + S[1, Seq[1]+1] + p[Seq[1]] 
        if t >= t_k
            Tardiness = max(0, t-d[Seq[1]])
            Reward = max(0.0, e[Seq[1]] - Tardiness*w[Seq[1]])
            if Reward > 0
                R[1,t] = Reward
                M[1,t] = Seq[1]
            end
        end
    end

    for k=2:n
        for t=1:T
            t_k = r[Seq[k]] + S[M[k,t]+1, Seq[k]+1] + p[Seq[k]] 
            if t >= t_k
                Tardiness = max(0, t-d[Seq[k]])
                Reward = max(0.0, e[Seq[k]] - Tardiness*w[Seq[k]])
                start_time = t - S[M[k,t]+1, Seq[k]+1] - p[Seq[k]]
                if R[k-1, start_time] + Reward > R[k-1, t]
                    R[k,t] = R[k-1, start_time] + Reward
                    M[k,t] = Seq[k]
                    P[k,t] = start_time
                else
                    R[k,t] = R[k-1,t]
                    M[k,t] = M[k-1,t]
                    P[k,t] = t
                end
            else
                R[k,t] = R[k-1,t]
                M[k,t] = M[k-1,t]
                P[k,t] = t
            end
        end
    end

    solution = Int[]
    prev_job = 0
    optimal_combo = argmax(R)
    k = optimal_combo[1]
    t = optimal_combo[2]
    obj = R[k,t]
    while true
        current_job = M[k,t]
        if current_job == 0
            break
        end
        if current_job != prev_job 
            push!(solution, current_job)
            prev_job = current_job
        end
        t = P[k,t]
        k = k-1 
        if t==0
            break
        end
    end
    return reverse(solution), obj
end

```
